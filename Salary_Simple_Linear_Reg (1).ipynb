**Fitting a Linear Regression Model**

```python
import statsmodels.formula.api as smf

model=smf.ols("Salary~YearsExperience",data=df).fit()
```

```python
sns.regplot(x="YearsExperience",y="Salary",data=df)
```

```python
Bo,B1=model.params
print("Interecept,Bo:",B1)
print("YearsExperience,B1:",B1)

#Getting p-value and t-value
tvalue,pvalue=(model.tvalues,model.pvalues)
print("t value:",tvalue)
print("pvalue:",pvalue)

#Getting R squared values
r_sq,r_sq_adj=(model.rsquared,model.rsquared_adj)
print("R2:",r_sq)
print("r2adj:",r_sq_adj)
```

```python
new_df=pd.Series([4,5])
new_df
```

```python
df_pred=pd.DataFrame(new_df,columns=['YearsExperience'])
df_pred
```

```python
model.predict(df_pred)
```

```python
predicted_salary = model.predict(df.YearsExperience)
pred_df = pd.DataFrame({'Salary_Actual' : df.Salary, 'Salary_Predicted' : predicted_salary})
pred_df
```

```python
# Plotting Actual Vs. Predicted Values
plt.scatter(df.YearsExperience, df.Salary,  color='green')
plt.plot(df.YearsExperience, predicted_salary, color='blue', linewidth=2)
plt.title('Actual Vs. Predicted Values')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error',metrics.mean_absolute_error(df.Salary,predicted_salary))
print("Mean Squared Error:",metrics.mean_squared_error(df.Salary,predicted_salary))
print("Root Mean square error:",np.sqrt(metrics.mean_squared_error(df.Salary,predicted_salary)))
print("R^ 2 mean:",metrics.r2_score(df.Salary,predicted_salary))
```

```python
model.summary()
```

**Building model with Logarithmic Transformation**

```python
data1 = pd.DataFrame.copy(df)
data1['log_YearsExperience'] = np.log(data1.YearsExperience)
data1.head()
```

```python
log_model = smf.ols('Salary ~ log_YearsExperience', data = data1).fit()
```

```python
# Coefficients Bo and B1
Bo, B1 = log_model.params
print("Intercept, Bo: ", Bo)
print("YearsExperience, B1: ", B1)

# Getting tvalue and pvalue
tvalue, pvalue = (log_model.tvalues, log_model.pvalues)
print("tvalue: ", tvalue)
print("pvalue: ", pvalue)

# Getting R squared values
r_sq, r_sq_adj = (log_model.rsquared, log_model.rsquared_adj)
print("R2: ", r_sq)
print("R2_Adj: ", r_sq_adj)
```

```python
predicted_salary = log_model.predict(data1.log_YearsExperience)
pred_df = pd.DataFrame({'Salary_Actual' : df.Salary, 'Salary_Predicted' : predicted_salary})
pred_df
```

```python
# Plotting Actual Vs. Predicted Values
plt.scatter(data1.YearsExperience, data1.Salary,  color='gray')
plt.plot(data1.YearsExperience, predicted_salary, color='red', linewidth=2)
plt.title('Actual Vs. Predicted Values')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(data1.Salary, predicted_salary))
print('Mean Squared Error:', metrics.mean_squared_error(data1.Salary, predicted_salary))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(data1.Salary, predicted_salary)))
print("R^2 Score:", metrics.r2_score(data1.Salary,predicted_salary))
```

```python
log_model.summary()
```
