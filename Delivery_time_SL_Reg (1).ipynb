**Fitting Linear Regression Model**

```python
import statsmodels.formula.api as smf

sns.regplot(x="sorting_time",y="delivery_time",data=df)
```

```python
model = smf.ols('delivery_time ~ sorting_time', data = df).fit()
```

```python
Bo,B1=model.params
print("Interecept,Bo:",B1)
print("Sorting time,B1:",B1)

#Getting p-value and t-value
tvalue,pvalue=(model.tvalues,model.pvalues)
print("t value:",tvalue)
print("pvalue:",pvalue)

#Getting R squared values
r_sq,r_sq_adj=(model.rsquared,model.rsquared_adj)
print("R2:",r_sq)
print("r2adj:",r_sq_adj)
```

```python
predicted_time = model.predict(df.sorting_time)
pred_df = pd.DataFrame({'Delivery_Time_Actual' : df.delivery_time, 'Delivery_Time_Predicted' : predicted_time})
pred_df
```

```python
#Plotting Actual vs Predicted
plt.scatter(df.sorting_time,df.delivery_time,color='green')
plt.plot(df.sorting_time,predicted_time,color='red',linewidth=2)
plt.title('Actual Vs. predicted values')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(df.delivery_time, predicted_time))
print('Mean Squared Error:', metrics.mean_squared_error(df.delivery_time, predicted_time))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(df.delivery_time, predicted_time)))
print("R^2 Score :", metrics.r2_score(df.delivery_time, predicted_time))
```

```python
model.summary()
```

**Building model with Logarithmic Transformation**

```python
# Preparing data for model building
data1 = pd.DataFrame.copy(df)
data1['log_sorting_time'] = np.log(data1.sorting_time)
```

```python
data1
```

```python
#Building Model
log_model = smf.ols('delivery_time~log_sorting_time' , data = data1).fit()
```

```python
# Coefficients Bo and B1
Bo, B1 = log_model.params
print("Intercept, Bo: ", Bo)
print("Sorting Time, B1: ", B1)

# Getting tvalue and pvalue
tvalue, pvalue = (log_model.tvalues, log_model.pvalues)
print("tvalue: ", tvalue)
print("pvalue: ", pvalue)

# Getting R squared values
r_sq, r_sq_adj = (log_model.rsquared, log_model.rsquared_adj)
print("R2: ", r_sq)
print("R2_Adj: ", r_sq_adj)
```

```python
predicted_time = log_model.predict(data1.log_sorting_time)
pred_df = pd.DataFrame({'Delivery_Time_Actual' : data1.delivery_time, 'Delivery_Time_Predicted' : predicted_time})
pred_df
```

```python
# Plotting Actual Vs. Predicted Values
plt.scatter(data1.log_sorting_time, data1.delivery_time,  color='red')
plt.plot(data1.log_sorting_time, predicted_time, color='blue', linewidth=2)
plt.title('Actual Vs. Predicted Values')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

```python

print('Mean Absolute Error:', metrics.mean_absolute_error(data1.delivery_time, predicted_time))
print('Mean Squared Error:', metrics.mean_squared_error(data1.delivery_time, predicted_time))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(data1.delivery_time, predicted_time)))
print("R^2 Score:", metrics.r2_score(data1.delivery_time, predicted_time))
```

```python
log_model.summary()
```
